<h2>1.1 Project Description</h2>

<h3>Goal/Purpose:</h3>

<p>
This project is an Exploratory Data Analysis (EDA) and Statistical Analysis of Hospital Emergency Room (ER) data. 
The goal is to analyze patient flow, waiting times, peak hours, and demographic patterns to understand 
Emergency Room congestion and provide data-driven recommendations to improve operational efficiency.
</p>

<p>
The purpose of this project is to simulate a real-world healthcare analytics case where hospital management 
requires insights about patient load, waiting time behavior, peak crowd hours, and service efficiency.
</p>

<p>
This project can serve as a professional portfolio project for Data Analyst and Healthcare Analytics roles.
</p>

<br>

<h3>Questions to be Answered:</h3>

<ul>
<li>Question No 01: What is the average waiting time in the Emergency Room?</li>
<li>Question No 02: Which hour of the day is the busiest?</li>
<li>Question No 03: Which day of the week has the highest patient load?</li>
<li>Question No 04: Which month has the most ER visits?</li>
<li>Question No 05: Do elderly patients wait longer than younger patients?</li>
<li>Question No 06: Does waiting time increase during peak hours?</li>
<li>Question No 07: Which triage level experiences the longest waiting time?</li>
<li>Question No 08: What is the distribution of patient age groups?</li>
<li>Question No 09: Are weekends more crowded than weekdays?</li>
<li>Question No 10: What operational improvements can reduce ER congestion?</li>
</ul>

<br>

<h3>Assumptions:</h3>

<ul>
<li>The dataset accurately records patient arrival time and waiting time.</li>
<li>Date and time columns are correctly formatted.</li>
<li>Waiting time is measured consistently across all patients.</li>
<li>Triage levels are correctly categorized.</li>
</ul>

<br>

<h3>Methodology:</h3>

<ul>
<li>Perform data cleaning (handle missing values, remove duplicates, correct date-time formats).</li>
<li>Create new features such as Hour of Arrival and Day of Week.</li>
<li>Perform univariate analysis (waiting time distribution, age distribution).</li>
<li>Perform peak time analysis (hourly, daily, monthly patient flow).</li>
<li>Compare waiting time across triage levels and age groups.</li>
<li>Perform correlation and heatmap analysis.</li>
<li>Create professional visualizations (bar charts, line charts, histograms, heatmaps).</li>
<li>Generate operational recommendations based on findings.</li>
</ul>

<br>

<h3>Scope:</h3>

<ul>
<li>This project focuses only on Emergency Room data analysis.</li>
<li>No machine learning prediction model is built (analysis-focused project).</li>
<li>The goal is to provide actionable hospital management insights.</li>
</ul>
