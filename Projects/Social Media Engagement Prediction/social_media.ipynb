<h2>1.1 Project Description</h2>

<h3>Goal/Purpose:</h3>

<p>
This project is an Exploratory Data Analysis (EDA) and Machine Learning Analysis of Social Media Engagement data. 
The goal is to analyze post characteristics, user behavior, platform trends, sentiment, and campaign features to predict 
engagement rates and provide data-driven recommendations for social media strategy optimization.
</p>

<p>
The purpose of this project is to simulate a real-world social media analytics case where marketers and content managers 
require insights about engagement drivers, content performance, audience behavior, and campaign effectiveness.

<p>

<h3>Questions to be Answered:</h3>

<ul>
<li>Question No 01: What factors most influence engagement rate?</li>

<li>Question No 03: Does day of the week impact engagement?</li>
<li>Question No 04: Which content types or topics receive the highest engagement?</li>
<li>Question No 05: How does sentiment or emotion of posts relate to engagement?</li>
<li>Question No 06: Do hashtags, mentions, or keywords increase engagement?</li>
<li>Question No 07: Can user history predict future engagement?</li>
<li>Question No 08: Are there patterns in engagement across campaigns or brands?</li>
<li>Question No 09: How does toxicity or negative sentiment affect engagement?</li>
<li>Question No 10: Which machine learning model best predicts engagement rates?</li>
</ul>

```python
df.isnull().sum()
```

<div style="background-color:#E6F2FF; padding: 10px; border-radius: 8px; box-shadow: 0 1px 3px 0 rgba(0, 0, 0, 0.1); border:1px solid #66C2A5; margin-top: 10px; margin-left: 20px;">
    <h2 style="font-size:20px; font-family:Georgia, serif; color:#333333; text-align: left; margin: 0;">
        5.2 Remove Outliers
    </h2>
</div>

```python
num_cols = df.select_dtypes(include=['int64','float64']).columns

for col in num_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    df = df[(df[col] >= lower) & (df[col] <= upper)]
```

```python
df.columns
```

<div style="background-color:#E6F2FF; padding: 10px; border-radius: 8px; box-shadow: 0 1px 3px 0 rgba(0, 0, 0, 0.1); border:1px solid #66C2A5; margin-top: 10px; margin-left: 20px;">

```python
drop_cols = ['post_id','timestamp','user_id','text_content',
             'hashtags','mentions','keywords',
             'brand_name','product_name','campaign_name']

df = df.drop(columns=drop_cols)
```

```python
cat_cols = ['platform','day_of_week','location','language',
            'topic_category','sentiment_label','emotion_type',
            'campaign_phase']
```

```python
df = pd.get_dummies(df, columns=cat_cols, drop_first=True)
```
