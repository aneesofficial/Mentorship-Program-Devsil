<h2>1.1 Project Description</h2>

<h3>Goal/Purpose:</h3>
<p>
This project aims to build a machine learning system that predicts healthcare costs and segments patients into meaningful groups. The project demonstrates handling missing data, encoding categorical variables, building and comparing regression models, applying regularization techniques, and performing clustering for patient segmentation.
</p>

<p>
The purpose of this notebook is to guide through the complete workflow of healthcare cost prediction and patient segmentation, including data preprocessing, regression modeling, regularization, clustering, model evaluation, and actionable insights generation.
</p>

<p>
This project can serve as a professional portfolio project for Data Analyst, Machine Learning, and Healthcare Analytics roles.
</p>

<br>

<h3>Tasks / Questions to be Answered:</h3>
<ul>
<li>Question 01: How can healthcare costs be predicted using patient features?</li>
<li>Question 02: Which patients belong to high-risk or high-cost groups?</li>
<li>Question 03: How do categorical features influence healthcare cost?</li>
<li>Question 04: Which regression model predicts cost most accurately?</li>
<li>Question 05: How does regularization (Ridge/Lasso) improve prediction?</li>
<li>Question 06: What is the optimal number of clusters for patient segmentation?</li>
<li>Question 07: How do clustering results help in real-world healthcare decisions?</li>
<li>Question 08: Which factors most affect healthcare cost?</li>
<li>Question 09: How do polynomial and multiple regression compare to linear regression?</li>
<li>Question 10: What actionable insights can be derived for high-cost patients?</li>
</ul>

```python
# Train test split
X_train, X_test, y_train, y_test = train_test_split(df.drop('charges', axis=1), df['charges'], test_size=0.2, random_state=42)
# Train a linear regression model
lr = LinearRegression()
lr.fit(X_train, y_train)

y_pred = lr.predict(X_test)

mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print(f"MAE: {mae}")
print(f"MSE: {mse}")
print(f"R2 Score: {r2}")

# Plot the scatter plot
plt.figure(figsize=(12, 6))
plt.scatter(y_test, y_pred, alpha=0.5)
plt.xlabel('Actual Charges')
plt.ylabel('Predicted Charges')
plt.title('Actual vs Predicted Charges')
plt.show()  

```

```python
# Polynomial Features
poly = PolynomialFeatures(degree=2)

X_poly = poly.fit_transform(df.drop('charges', axis=1))
y = df['charges']

# train test split
X_train_poly, X_test_poly, y_train, y_test = train_test_split(X_poly, y, test_size=0.2, random_state=42)

# Train a linear regression model
poly_model = LinearRegression()
poly_model.fit(X_train_poly, y_train)
y_pred_poly = poly_model.predict(X_test_poly)

mae_poly = mean_absolute_error(y_test, y_pred_poly)
mse_poly = mean_squared_error(y_test, y_pred_poly)
r2_poly = r2_score(y_test, y_pred_poly)
print(f"Polynomial MAE: {mae_poly}")
print(f"Polynomial MSE: {mse_poly}")
print(f"Polynomial R2 Score: {r2_poly}")

```

```python
# Apply Kmeans Clustering
kmeans = KMeans(n_clusters=3, random_state=42)
df['cluster'] = kmeans.fit_predict(df.drop('charges', axis=1))
df.head()
```

```python
# Determine Optimal Number of Clusters (Elbow Method)
inertia = []

for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(df.drop(['charges', 'cluster'], axis=1))
    inertia.append(kmeans.inertia_)

plt.plot(range(1, 11), inertia, marker="o")
plt.title("Elbow Method")
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Inertia")
plt.show()
```
